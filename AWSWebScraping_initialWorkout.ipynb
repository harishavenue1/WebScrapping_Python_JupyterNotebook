```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.amazon.in/stores/HPLaptops/page/DC0D6D54-3324-4C73-9E66-863A76269836?ref_=ast_bln"
```

```python
#headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.3', 'Accept-Language':'en-US'})
```

```python
webpage = requests.get(url, headers=HEADERS)
```

```python
webpage
```

```python
webpage.content
```

```python
type(webpage.content)
```

```python
#convert the webpage bytes format to html tag format
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
# get all the products links on the screen
links = soup.find_all("a", attrs={'class':'Overlay__overlay__LloCU ProductGridItem__overlay__IQ3Kw'})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
product_list = "https://www.amazon.in/" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
#convert the webpage bytes format to html tag format
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
# get all the products details on the screen
title = new_soup.find("span", attrs={'id':'productTitle'}).text.strip()
```

```python
# get all the products details on the screen
price = new_soup.find("span", attrs={'class':'a-offscreen'}).text.strip()
price
```

```python
# get all the products ratings on the screen
ratings = new_soup.find("span", attrs={'class':'a-icon-alt'}).text.strip()
ratings
```

```python
# get all the products availability on the screen
ratings = new_soup.find("div", attrs={'id':'availability'}).text.strip()
ratings
```

```python
# reviews
reviews = new_soup.find("span", attrs={'id':'acrCustomerReviewText'}).text.strip()
reviews
```
